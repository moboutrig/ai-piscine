````python
!pip3 install pandas numpy

import pandas as pd
from pandas import Series, DataFrame
import numpy as np

# init
df = pd.read_csv("household_power_consumption.txt", sep=";")

# 1.
df = df.drop(['Time', 'Sub_metering_2', 'Sub_metering_3'], axis=1)
df['Date'] = pd.to_datetime(df['Date'])

# 2.
df = df.set_index('Date')

# testing
# print(df.head().index)

# 3.
# if 'coerce', then invalid parsing will be set as NaN.
def update_types(df) :
    for c in df:
        df[c] = pd.to_numeric(df[c], errors='coerce')
    return df

df = update_types(df)

# testing
# print(df.dtypes)

# 4.
print(df.describe())

# testing with NaN's
# print(df.isna().sum())

# 5.
df.dropna(inplace=True)

# testing witout NaN's
# print(df.isna().sum())

# 6.
df.loc[:,'Sub_metering_1'] = df['Sub_metering_1'] * 0.06

# 7. Select all the rows for which the Date is greater than 2008-12-27 and `Voltage` is greater than 242
# color_and_shape = df.loc[(df['Color'] == 'Green') & (df['Shape'] == 'Rectangle')]
filtered_df = df.loc[(df.index >= '2008-12-27') & (df['Voltage'] >= 242)]

# test
# print(filtered_df.head().to_markdown())
# print(filtered_df.count())

# 8. Print the 88888th row.
# print("-------------")
df['index'] = np.arange(len(df))
print(df.loc[df['index'] == 88888])

# 9. What is the date for which the `Global_active_power` is maximal
# print('--------------')
print(df[['Global_active_power']].idxmax())

# 10. Sort the first three columns by descending order of `Global_active_power` and ascending order of `Voltage`.
sorted_df = df.sort_values(by = ['Global_active_power', 'Voltage'], ascending = [False, True])
sorted_df = sorted_df.drop(['Global_intensity', 'Sub_metering_1', 'index'], axis=1)

# test
# print('--------------')
# print(sorted_df.tail().to_markdown())

# 11. Compute the daily average of `Global_active_power`.

means = df.groupby(pd.Grouper(freq='1D'))['Global_active_power'].mean()

print(means)
````
