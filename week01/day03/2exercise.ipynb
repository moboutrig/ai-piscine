````python
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame({
'name':['christopher','marion','maria','mia','clement','randy','remi'],
'age':[70,30,22,19,45,33,20],
'gender':['M','F','F','F','M','M','M'],
'state':['california','dc','california','dc','california','new york','porto'],
'num_children':[2,0,0,3,8,1,4],
'num_pets':[5,1,0,5,2,2,3]
})

print(df)

plt.yticks(np.arange(0, 4.5, 0.5))
# change the value 8 because its hard to have 8 childs so lets say that this value is
# a mising value so we must use the median to change the value for samething better
df.loc[4, 'num_children'] = df['num_children'].median()
plt.scatter(df['age'].sort_values(ascending=True), df['num_children'].sort_values(ascending=True))

plt.xlabel('age')
plt.ylabel('num_children')
plt.title('Scatter plot: Number of children and age')
plt.show()
````
