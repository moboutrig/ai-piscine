The goal of this exercise is to learn to compute and plot the sigmoid function.

1. On the same plot, plot the sigmoid function and the custom sigmoids defined as:

```
  - sigmoid1(x) = 1/(1+ exp(-(0.5*x + 3)))
  - sigmoid2(x) = 1/(1+ exp(-(5*x + 11)))
```

- Add a line representing the probability=0.5

````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression

# sigmoid1(x) = 1/(1+ exp(-(0.5*x + 3)))
# sigmoid2(x) = 1/(1+ exp(-(5*x + 11)))
x = np.arange(-20, 15)
y = np.arange(-20.0, 15.0)
y.fill(0.5)
sigmoid1= 1 / (1 + np.exp(-(0.5 * x + 3)))
sigmoid2= 1 / (1 + np.exp(-(5 * x + 11)))
sigmoid3= 1 / (1 + np.exp(-(x + 0.5)))
plt.plot(x, y, color='red')
plt.plot(x, sigmoid1, color='orange')
plt.plot(x, sigmoid2, color='green')
plt.plot(x, sigmoid3, color='blue')
plt.title('Sigmoids')
plt.xlabel('x')
plt.show()
````
